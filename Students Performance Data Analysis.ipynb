# Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
df = pd.read_csv("student_scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Remove Unnamed column

```python
df = df.drop("Unnamed: 0", axis = 1)
df.head()
```

```python
df.tail()
```

# Change Weekly Study Hour Column

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-oct","5-10")
df.head()
```

# Gender distribution

```python
plt.figure(figsize = (4,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

#### The number of females in the data are more than the number of males

# Parent's Education Impact On Student's Performance

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize = (3,5))
sns.heatmap(gb,annot = True)
plt.title("Parent's Education Vs Student's Performance")
plt.show()
```

#### The education of the parents have a good impact on their children.
#### The average scores of children whose parents hold master's degrees are notably higher than those of children whose parents graduated from high school.

# Parent's Marital Status Impact On Student's Performance

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize = (3,5))
sns.heatmap(gb1,annot = True)
plt.title("Parent's Merital Status Vs Student's Performance")
plt.show()
```

#### Marital status among parents does not significantly impact students' scores.

# Distribution of Ethnic Group

```python
print(df["EthnicGroup"].unique())
```

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

groups = ["groupA", "groupB", "groupC", "groupD", "groupE"]
groups_count = [groupA['EthnicGroup'], groupB['EthnicGroup'], groupC['EthnicGroup'], groupD['EthnicGroup'], groupE['EthnicGroup']]

plt.pie(groups_count, labels = groups, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```

```python
ax = sns.countplot(data = df, x = 'EthnicGroup')
plt.title("Ethnic groups vs count of students")
ax.bar_label(ax.containers[0])
```

#### The majority of students belong to group C, while group A has the fewest students.

# Sports Participation Analysis

```python
ax = sns.countplot(data = df, x = 'PracticeSport')
plt.title("Sports Participation vs Count")
ax.bar_label(ax.containers[0])
```

```python
gb2 = df.groupby("PracticeSport").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb2)
```

```python
plt.figure(figsize = (3,5))
sns.heatmap(gb2,annot = True)
plt.title("Sports Participation Vs Student's Performance")
plt.show()
```

#### Students who do not participate in sports tend to have the lowest performance, while those who regularly participate in sports exhibit the best performance.
